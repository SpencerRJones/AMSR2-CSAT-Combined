```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import glob
```

```python
#---------------------------------------------
#
#   Show Tb biases with radar-only retrievals 
#
#---------------------------------------------

data_dir = '/edata2/spencer/OE/amsr2_radar/outputs/retrievals/'

beg_date = '2015-01-01'
end_date = '2015-01-31'

beg_date = np.datetime64(beg_date)
end_date = np.datetime64(end_date)

first = True


for idate in np.arange(beg_date, end_date + np.timedelta64(1, "D")):
    
    print(idate)
    
    fdate = f"{str(idate).replace('-', '')}"
    
    
    flist = glob.glob(f"{data_dir}{fdate[:-2]}/{fdate}/*_cloudy_radaronly.nc")
    flist.sort()
    
    if len(flist) == 0:
        continue
    
    for ifile in flist:
        
        print(ifile)
        
        with xr.open_dataset(ifile) as f:
            tbobs = f.TbsObs.values.T
            tbsim = f.TbsSim.values.T
            
        good = np.where(np.all(tbsim > 0., axis=1))
        
        tbobs = tbobs[good]
        tbsim = tbsim[good]
        
        if first:
            all_tbobs = tbobs
            all_tbsim = tbsim
            first = False
            
        else:
            all_tbobs = np.concatenate((all_tbobs,tbobs), axis=0)
            all_tbsim = np.concatenate((all_tbsim,tbsim), axis=0)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=5, figsize=[16,7], tight_layout=True)

ichan = 0

one2one = np.arange(-1000,1000)


for irow in np.arange(0,2):
    for icol in np.arange(0,5):
        ax[irow,icol].plot(one2one, one2one, linestyle='--',linewidth=2)
        
#         ichan += 1
        
ax1  = ax[0,0]
ax2  = ax[0,1]
ax3  = ax[0,2]
ax4  = ax[0,3]
ax5  = ax[0,4]
ax6  = ax[1,0]
ax7  = ax[1,1]
ax8  = ax[1,2]
ax9  = ax[1,3]
ax10 = ax[1,4]

lim1_10v = 160.
lim2_10v = 175.

lim1_10h = 75.
lim2_10h = 95.

lim1_19v = 180.
lim2_19v = 195.

lim1_19h = 95
lim2_19h = 130

lim1_23v = 190
lim2_23v = 230

lim1_23h = 125
lim2_23h = 180

lim1_37v = 120
lim2_37v = 200

lim1_37h = 205
lim2_37h = 220

lim1_89v = 125
lim2_89v = 160

lim1_89h = 235
lim2_89h = 265

nbins = 50

ax1.hist2d(all_tbsim[:,0], all_tbobs[:,0], bins=nbins, 
           range=[[lim1_10v,lim2_10v],[lim1_10v,lim2_10v]], 
           facecolor='k', cmap='magma')
ax2.hist2d(all_tbsim[:,1], all_tbobs[:,1], bins=nbins, 
           range=[[lim1_10h,lim2_10h],[lim1_10h,lim2_10h]], 
           facecolor='k', cmap='magma')
ax3.hist2d(all_tbsim[:,2], all_tbobs[:,2], bins=nbins, 
           range=[[lim1_19v,lim2_19v],[lim1_19v,lim2_19v]], 
           facecolor='k', cmap='magma')
ax4.hist2d(all_tbsim[:,3], all_tbobs[:,3], bins=nbins, 
           range=[[lim1_19h,lim2_19h],[lim1_19h,lim2_19h]], 
           facecolor='k', cmap='magma')
ax5.hist2d(all_tbsim[:,4], all_tbobs[:,4], bins=nbins, 
           range=[[lim1_23v,lim2_23v],[lim1_23v,lim2_23v]], 
           facecolor='k', cmap='magma')
ax6.hist2d(all_tbsim[:,5], all_tbobs[:,5], bins=nbins, 
           range=[[lim1_23h,lim2_23h],[lim1_23h,lim2_23h]], 
           facecolor='k', cmap='magma')
ax7.hist2d(all_tbsim[:,5], all_tbobs[:,5], bins=nbins, 
           range=[[lim1_37v,lim2_37v],[lim1_37v,lim2_37v]], 
           facecolor='k', cmap='magma')
ax8.hist2d(all_tbsim[:,6], all_tbobs[:,6], bins=nbins, 
           range=[[lim1_37h,lim2_37h],[lim1_37h,lim2_37h]], 
           facecolor='k', cmap='magma')
ax9.hist2d(all_tbsim[:,7], all_tbobs[:,7], bins=nbins, 
           range=[[lim1_89v,lim2_89v],[lim1_89v,lim2_89v]], 
           facecolor='k', cmap='magma')
ax10.hist2d(all_tbsim[:,8], all_tbobs[:,8], bins=nbins, 
           range=[[lim1_89h,lim2_89h],[lim1_89h,lim2_89h]], 
           facecolor='k', cmap='magma')


# ax1.set_xlim(162,175)
# ax1.set_ylim(162,175)
# ax2.set_xlim(80,95)
# ax2.set_ylim(80,95)
# ax3.set_xlim(180,198)
# ax3.set_ylim(180,198)
# ax4.set_ylim(180,126)
# ax5.set_ylim(213,222)
# ax6.set_ylim(153,168)
# ax7.set_ylim(215,231)
# ax8.set_ylim(145,180)
# ax9.set_ylim(257,273)
# ax10.set_ylim(220,265)

ax1.set_title('10V', fontsize=14)
ax2.set_title('10H', fontsize=14)
ax3.set_title('19V', fontsize=14)
ax4.set_title('19H', fontsize=14)
ax5.set_title('23V', fontsize=14)
ax6.set_title('23H', fontsize=14)
ax7.set_title('37V', fontsize=14)
ax8.set_title('37H', fontsize=14)
ax9.set_title('89V', fontsize=14)
ax10.set_title('89H', fontsize=14)

ax1.set_ylabel('Observed Tb [K]', fontsize=14)
ax6.set_ylabel('Observed Tb [K]', fontsize=14)

ax6.set_xlabel('Simulated Tb [K]', fontsize=14)
ax7.set_xlabel('Simulated Tb [K]', fontsize=14)
ax8.set_xlabel('Simulated Tb [K]', fontsize=14)
ax9.set_xlabel('Simulated Tb [K]', fontsize=14)
ax10.set_xlabel('Simulated Tb [K]', fontsize=14)

ax1.set_yticks(ax1.get_xticks())
ax2.set_yticks(ax2.get_xticks())
ax3.set_yticks(ax3.get_xticks())
ax4.set_yticks(np.arange(95,130,10))
ax4.set_xticks(np.arange(95,130,10))
ax5.set_yticks(ax5.get_xticks())
ax6.set_xticks(np.arange(lim1_23h,lim2_23h,10))
ax6.set_yticks(np.arange(lim1_23h,lim2_23h,10))
ax7.set_yticks(ax7.get_xticks())
ax8.set_yticks(ax8.get_xticks())
ax9.set_yticks(np.arange(lim1_89v,lim2_89v,10))
ax9.set_xticks(np.arange(lim1_89v,lim2_89v,10))
ax10.set_xticks(np.arange(lim1_89h,lim2_89h+5,5))
ax10.set_yticks(np.arange(lim1_89h,lim2_89h+5,5))

#ax1.legend(fontsize=12)
plt.suptitle('AMSR2 Simulated Tbs from CPR Profiles.', fontsize=18)
plt.show()
```
