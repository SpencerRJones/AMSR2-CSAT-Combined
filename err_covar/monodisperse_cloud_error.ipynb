```python
'''

'''

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
```

```python
infile1 = '/edata2/spencer/thesis_data/monodisperse_cloud_diam_err/monodisperse_cloud_Tbs_clwc0.05.bin'
infile2 = '/edata2/spencer/thesis_data/monodisperse_cloud_diam_err/monodisperse_cloud_Tbs_clwc0.10.bin'
infile3 = '/edata2/spencer/thesis_data/monodisperse_cloud_diam_err/monodisperse_cloud_Tbs_clwc0.25.bin'

nd = 96
nchans = 10

tbs = np.zeros([nd, nchans], dtype='f')

with open(infile1,'rb') as f:
    tbs1 = np.fromfile(f, sep='', count=nd*nchans, dtype='f').reshape(nchans, nd).T
    
with open(infile2, 'rb') as f:
    tbs2 = np.fromfile(f, sep='', count=nd*nchans, dtype='f').reshape(nchans, nd).T
    
with open(infile3, 'rb') as f:
    tbs3 = np.fromfile(f, sep='', count=nd*nchans, dtype='f').reshape(nchans, nd).T
```

```python
tbs.shape
```

```python
D = np.zeros([nd])
for i in np.arange(0,nd):
    D[i] = 0.005 + (0.001 * i)
    
D
```

```python
fig, ax = plt.subplots(nrows=2, ncols=5, figsize=[16,7], sharex=True)

ichan = 0
for irow in np.arange(0,2):
    for icol in np.arange(0,5):
        ax[irow,icol].plot(D, tbs1[:,ichan], linewidth=2.5, label='Dry Cloud')
        ax[irow,icol].plot(D, tbs2[:,ichan], linewidth=2.5, label='Medium Cloud')
        ax[irow,icol].plot(D, tbs3[:,ichan], linewidth=2.5, label='Juicy Cloud')
        ichan += 1
        
ax1  = ax[0,0]
ax2  = ax[0,1]
ax3  = ax[0,2]
ax4  = ax[0,3]
ax5  = ax[0,4]
ax6  = ax[1,0]
ax7  = ax[1,1]
ax8  = ax[1,2]
ax9  = ax[1,3]
ax10 = ax[1,4]

ax1.set_ylim(171,174)
ax2.set_ylim(85,90)
ax3.set_ylim(190,198)
ax4.set_ylim(112,126)
ax5.set_ylim(213,222)
ax6.set_ylim(153,168)
ax7.set_ylim(215,231)
ax8.set_ylim(145,180)
ax9.set_ylim(257,273)
ax10.set_ylim(220,265)

ax1.set_title('10V', fontsize=14)
ax2.set_title('10H', fontsize=14)
ax3.set_title('19V', fontsize=14)
ax4.set_title('19H', fontsize=14)
ax5.set_title('23V', fontsize=14)
ax6.set_title('23H', fontsize=14)
ax7.set_title('37V', fontsize=14)
ax8.set_title('37H', fontsize=14)
ax9.set_title('89V', fontsize=14)
ax10.set_title('89V', fontsize=14)

ax1.set_ylabel('Tb [K]', fontsize=14)
ax6.set_ylabel('Tb [K]', fontsize=14)

ax6.set_xlabel('D [mm]', fontsize=14)
ax7.set_xlabel('D [mm]', fontsize=14)
ax8.set_xlabel('D [mm]', fontsize=14)
ax9.set_xlabel('D [mm]', fontsize=14)
ax10.set_xlabel('D [mm]', fontsize=14)

ax1.legend(fontsize=12)
plt.show()
```
